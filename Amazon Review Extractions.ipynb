```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_soup(u):
    r = requests.get(u,headers=HEADERS)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup
review_list=[]
```

```python
def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {
            'title': item.find('a', {'data-hook': 'review-title'}).text.strip(),
            'rating':float(item.find('i', {"data-hook":"review-star-rating"}).text.replace('out of 5 stars', '').strip()),
            'body': item.find('span', {'data-hook':"review-body"}).text.strip(),
            }
            review_list.append(review)
    except:
        pass
```

```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
            AppleWebKit/537.36 (KHTML, like Gecko) \
            Chrome/90.0.4430.212 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
```

```python
for x in range(0,21):
    soup = get_soup(f'https://www.amazon.com/Santamedical-Oximeter-Fingertip-Saturation-Batteries/product-reviews/B08HH86RT2/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews{x}')
    #print(soup)
    print(f'Getting page: {x}')
    get_reviews(soup)
    print(len(review_list))
    if not soup.find('li', {'class': 'a-disabled a-last'}):
        pass
    else:
        break
```

```python
final_list = pd.DataFrame(review_list)
final_list
```

```python
final_list.to_csv(r'Oximeter Reviews.csv', index=False)
final_list
```
